```python
#converting timestamp column to a datetime feature
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
#separating dayofweek from timestamp
df['Dayofweek'] = df['Timestamp'].dt.dayofweek
df['Dayofweek'].head()
```

```python
#removing the spaces in column names and adding underscores
df.rename(columns=({'Clicked on Ad':'Clicked_on_Ad','Daily Time Spent on Site' :'Daily_Time_Spent_on_Site','Area Income':'Area_Income','Daily Internet Usage':'Daily_Internet_Usage','Ad Topic Line':'Ad_Topic_Line'}),inplace=True)
```

```python
def outlier(feature):
        stat = df[feature].describe()
        iqr = stat['75%']-stat['25%']
        upper = stat['75%'] + 1.5*iqr
        lower = stat['25%'] - 1.5*iqr
        print('Upper is {} '.format(upper))
        print('Lower is {} '.format(lower))
```

# Baseline Model

### Creating a baseline model

```python
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#initiating the features and target variables
X = df[['Daily_Time_Spent_on_Site', 'Age', 'Area_Income', 'Daily_Internet_Usage', 'Male']]
y = df['Clicked_on_Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42) 
```

```python
#create an instance of logistic regression 
lr = LogisticRegression()
```

```python
#training the model with training data
lr.fit(X_train,y_train)
```

```python
#predicting for testing dataset
pred = lr.predict(X_test)
```

```python
# Printing classification_report to see the results
print(classification_report(y_test, pred))
```

```python
# Printing the confusion_matrix
print(confusion_matrix(y_test, pred))
```

```python
print(accuracy_score(y_test,pred))
```

# Model

```python
#making a copy of the dataframe so that the original dataframe stays intact
df_cpy = df.copy()
```

```python
#creating dummy variables
df_cpy = pd.concat([df_cpy, pd.get_dummies(df_cpy['Country'])], axis=1)
df_cpy = pd.concat([df_cpy,pd.get_dummies(df_cpy['Dayofweek'])],axis=1)
df_cpy = pd.concat([df_cpy,pd.get_dummies(df_cpy['Month'])],axis=1)
```

```python
#creating bins for different hours in a day
df_cpy['Hour_bins'] = pd.cut(df_cpy['Hour'],bins = [0,5,11,17,23],labels = ['Midnight','Morning','Afternoon','Evening'],include_lowest=True)
```

```python
df_cpy = pd.concat([df_cpy,pd.get_dummies(df_cpy['Hour_bins'])],axis=1)
```

```python
df_cpy.drop(['Country', 'Ad_Topic_Line', 'City', 'Day', 'Month', 'Dayofweek', 'Hour', 'Hour_bins','Timestamp'],axis=1,inplace=True)
```

```python
df_cpy.head()
```

```python
#creating the feature and the target variable
df_cpy_features = df_cpy.drop(['Clicked_on_Ad'],axis = 1)
df_cpy_target = df_cpy['Clicked_on_Ad']
```

```python
#scaling the features 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_X_new = scaler.fit_transform(df_cpy_features)
scaled_X_new
```

```python
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(scaled_X_new, df_cpy_target, test_size=0.30, random_state=42) 
```

## Logistic Regression

```python
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train_new,y_train_new)
```

```python
#copying the predicted values of the test dataset
predictions = log_reg.predict(X_test_new)
```

```python
print(classification_report(y_test_new, predictions))
```

```python
print(confusion_matrix(y_test_new, predictions))
```

```python
print(accuracy_score(y_test_new,predictions))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=600,random_state=42,min_samples_split=10,
                                  min_samples_leaf=1,oob_score=True,max_depth=20)
```

```python
rf.fit(X_train_new,y_train_new)
```

```python
# Predict using model
rf_training_pred = rf.predict(X_test_new)
```

```python
print(confusion_matrix(y_test_new,rf_training_pred))
```

```python
print(accuracy_score(y_test_new,rf_training_pred))
```
